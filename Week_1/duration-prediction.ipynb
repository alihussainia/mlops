```python
# imports
import pandas as pd
```

```python
# use tab to autofill path
# reading parquet files
df = pd.read_parquet('./data/jan.parquet')
```

```python
# checking first five rows
df.head()
```

```python
# extracting only those rows with having trip_type value 2 i.e. dispatchs  
df = df[df.trip_type == 2]
```

```python
# checking first five rows 
df.head()
```

```python
# checking data types of columns
df.dtypes
```

```python
# we are interested in duration from pickup to dropoff
# creating duration column and then converting it into minutes
df['duration'] = df.lpep_dropoff_datetime - df.lpep_pickup_datetime
df.duration = df.duration.apply(lambda td: td.total_seconds()/ 60)
```
